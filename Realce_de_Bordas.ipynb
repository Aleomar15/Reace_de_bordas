```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
import matplotlib.pyplot as plt
import math

imagem =  cv2.imread("/content/Pessoa.jpg")
cv2_imshow(imagem)
```

```python
img_cinza = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_cinza)
```

```python
roberts1 = np.array([[1,0], [0,-1]])
op_roberts1 = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = roberts1)
cv2_imshow(op_roberts1)
```

```python
roberts2 = np.array([[0,1], [-1,0]])
op_roberts2 = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = roberts2)
cv2_imshow(op_roberts2)
```

```python
imagem_final_roberts =  np.zeros((img_cinza.shape[0], img_cinza.shape[1]))

for linha in range(img_cinza.shape[0]):
  for coluna in range(img_cinza.shape[1]):
    imagem_final_roberts[linha][coluna] = math.sqrt(pow(op_roberts1[linha][coluna],2) + pow(op_roberts2[linha][coluna],2))
cv2_imshow(imagem_final_roberts)
```

```python
sobel = np.array([[-1,-2,-1], [0,0,0], [1,2,1]])
op_sobel = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = sobel)
cv2_imshow(op_sobel)
```

```python
sobel2 = np.array([[1,2,1], [0,0,0], [-1,-2,-1]])
op_sobel2 = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = sobel2)
cv2_imshow(op_sobel2)
```

```python
imagem_final_sobel =  np.zeros((img_cinza.shape[0], img_cinza.shape[1]))

for linha in range(img_cinza.shape[0]):
  for coluna in range(img_cinza.shape[1]):
    imagem_final_sobel[linha][coluna] = op_sobel[linha][coluna] + op_sobel2[linha][coluna]
cv2_imshow(imagem_final_sobel)
```

```python
prewit = np.array([[-1,-1,-1], [0,0,0], [1,1,1]])
op_prewit = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = prewit)
cv2_imshow(op_prewit)
```

```python
prewit2 = np.array([[1,1,1], [0,0,0], [-1,-1,-1]])
op_prewit2 = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = prewit2)
cv2_imshow(op_prewit2)
```

```python
imagem_final_prewit =  np.zeros((img_cinza.shape[0], img_cinza.shape[1]))

for linha in range(img_cinza.shape[0]):
  for coluna in range(img_cinza.shape[1]):
    imagem_final_prewit[linha][coluna] = op_prewit[linha][coluna] + op_prewit2[linha][coluna]
cv2_imshow(imagem_final_prewit)
```

```python
imagem_final_laplaciano =  np.zeros((img_cinza.shape[0], img_cinza.shape[1]))

laplaciano = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
laplaciano = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = laplaciano)
cv2_imshow(laplaciano)
```

```python
imagem_final_pI =  np.zeros((img_cinza.shape[0], img_cinza.shape[1]))

pI = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]])
pI = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = pI)
cv2_imshow(pI)
```

```python
lIH = np.array([[-1,-1,-1], [2,2,2], [-1,-1,-1]])
lIH = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = lIH)
cv2_imshow(lIH)
```

```python
lIV = np.array([[-1,2,-1], [-1,2,-1], [-1,2,-1]])
lIV = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = lIV)
cv2_imshow(lIV)
```

```python
lI45p = np.array([[-1,-1,2], [-1,2,-1], [2,-1,-1]])
lI45p = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = lI45p)
cv2_imshow(lI45p)
```

```python
lI45n = np.array([[2,-1,-1], [-1,2,-1], [-1,-1,2]])
lI45n = cv2.filter2D(src = img_cinza, ddepth = -1, kernel = lI45n)
cv2_imshow(lI45n)
```

```python
imagem_final_lI =  np.zeros((img_cinza.shape[0], img_cinza.shape[1]))

for linha in range(img_cinza.shape[0]):
  for coluna in range(img_cinza.shape[1]):
    imagem_final_lI[linha][coluna] = lIH[linha][coluna] + lIV[linha][coluna]+ lI45p[linha][coluna] + lI45n[linha][coluna]
cv2_imshow(imagem_final_lI)
```
